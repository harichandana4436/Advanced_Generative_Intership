# Problem 1: Social Media – Post Engagement Analyzer
Problem Statement

Create a function to analyze engagement on a social media post.
    
Rules
    
Calculate total likes

If Total Likes ≥ 1000 → Viral Post

Otherwise → Normal Engagement

Real-Time Use

Social media analytics platforms

Marketing dashboards

Hint

Use function

Use loop for calculations

Expected Output Format

Total Likes: 1050

Post Status: Viral Post


```python
def analyze_post_engagement(likes):
    total_likes = 0

    # Loop to calculate total likes
    for like in likes:
        total_likes += like

    # Determine status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Output format
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")


# Example usage
likes_data = [200, 150, 300, 400]  # Total = 1050
analyze_post_engagement(likes_data)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker
Problem Statement

Create a function that analyzes seasonal rainfall data.
    
Rules

Calculate average rainfall

Average ≥ required level → Adequate Rainfall

Otherwise → Inadequate Rainfall

Real-Time Use

Climate monitoring systems

Irrigation planning tools

Hint

Use function

Use loop and math operations

Expected Output Format

Average Rainfall: 72

Rainfall Status: Adequate Rainfall


```python
def rainfall_adequacy_checker(rainfall_data, required_level):
    total = 0

    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total += rain

    # Calculate average rainfall
    average = total / len(rainfall_data)

    # Check adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    # Output format
    print(f"Average Rainfall: {int(average)}")
    print(f"Rainfall Status: {status}")


# Example usage
rainfall_data = [60, 75, 80, 70]
required_level = 70
rainfall_adequacy_checker(rainfall_data, required_level)
```

# Problem 4: Social Media – Duplicate Account Detection
Problem Statement

Create a function to detect duplicate usernames in a system.
    
Rules
    
If duplicates exist → Flag accounts

Else → Accounts are unique

Real-Time Use

Platform moderation

Fraud detection systems

Hint

Use function

Use set data structure

Expected Output Format

Duplicate Accounts Found: Yes


```python
def detect_duplicate_accounts(usernames):
    # Use set to track unique usernames
    seen = set()
    duplicates_found = False

    # Loop through usernames
    for name in usernames:
        if name in seen:
            duplicates_found = True
            break
        seen.add(name)

    # Output result
    if duplicates_found:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage
user_list = ["alice", "bob", "charlie", "alice"]
detect_duplicate_accounts(user_list)
```

# Problem 6: Agriculture – Premium Crop Price Filter
Problem Statement

Create a function that filters crop prices above a premium threshold.
    
Rules
    
Price > ₹2000 → Premium Crop

Else → Regular Crop

Real-Time Use

Market analytics

Export pricing systems

Hint
Use function

Use list filtering logic

Expected Output Format

Premium Crops: [2500, 3200]


```python
def premium_crop_filter(prices):
    premium_crops = []

    # Loop through prices and filter premium crops
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    # Output format
    print(f"Premium Crops: {premium_crops}")


# Example usage
crop_prices = [1500, 2500, 1800, 3200, 2000]
premium_crop_filter(crop_prices)
```

# Problem 9: E-Learning – Student Attendance Eligibility System
Problem Statement

Create a function to calculate attendance percentage and check exam eligibility.
    
Rules
    
Attendance ≥ 75% → Eligible

Attendance < 75% → Not Eligible

Real-Time Use

Learning management systems

Internship tracking platforms

Hint

Use function

Use list and loop

Expected Output Format

Attendance Percentage: 80.0

Exam Eligibility: Eligible


```python
def attendance_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Loop to count present days
    for status in attendance_list:
        if status == 1:   # 1 = Present, 0 = Absent
            present_days += 1

    # Calculate percentage
    percentage = (present_days / total_days) * 100

    # Check eligibility
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    # Output format
    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")


# Example usage
attendance_data = [1, 1, 1, 1, 0, 1, 1, 1, 0, 1]  # 8/10 = 80.0%
attendance_eligibility(attendance_data)
```
