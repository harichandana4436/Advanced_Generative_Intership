# Problem 1: Mobile Recharge Validation System
Problem Statement

Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be ≥ ₹50


Valid plans: 199, 299, 399, 599


User should be allowed to retry if invalid


Real-Time Use
Telecom recharge apps


Payment gateways


Hint
Use function for validation


Use while loop for retry


```python
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        return False, "Recharge amount must be at least ₹50."

    if plan not in valid_plans:
        return False, "Invalid plan selected. Choose 199, 299, 399, or 599."

    return True, "Recharge request is valid."

while True:
    try:
        amount = int(input("Enter recharge amount: "))
        plan = int(input("Enter plan (199/299/399/599): "))

        is_valid, message = validate_recharge(amount, plan)
        print(message)

        if is_valid:
            print("Recharge successful!")
            break
        else:
            print("Please try again...\n")

    except ValueError:
        print("Invalid input! Please enter numeric values only.\n")
```

# Problem 2: Inventory Reorder Alert System
Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15


Print Stock OK otherwise


Real-Time Use
Warehouse systems


Retail management


Hint
Use function


Use for loop on dictionary


```python
def check_stock(products):
    for item in products:
        if products[item] < 15:
            print(item, "Reorder Alert")
        else:
            print(item, "Stock OK")


# Example data
products = {
    "Pen": 10,
    "Book": 25,
    "Notebook": 5,
    "Eraser": 20
}

check_stock(products)
```

#  Problem 3: Student Result Processing System
Problem Statement

Create a function that:

Accepts a list of student marks


Calculates average marks


Prints Pass if average ≥ 50 else Fail


Real-Time Use
Online exam systems


Evaluation portals


Hint
Use loop to calculate total


Return result from function


```python
def result(marks):
    total = 0

    for mark in marks:
        total = total + mark

    avg = total / len(marks)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"


# Example
marks = [60, 70, 40, 80]
print(result(marks))
```

# Problem 4: Cab Fare Estimator with Retry Option
Problem Statement
Create a cab fare calculator:
Base fare = ₹50


₹12 per km


Peak hour → 25% extra


Allow retry using while loop


Real-Time Use
Ride-hailing apps


Transport systems


Hint
Function for fare calculation


While loop for retry


```python
def calculate_fare(km, peak):
    base_fare = 50
    fare = base_fare + (km * 12)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    km = float(input("Enter distance in km: "))
    peak = input("Peak hour? (yes/no): ")

    total = calculate_fare(km, peak)
    print("Total Fare:", total)

    retry = input("Calculate again? (yes/no): ")
    if retry != "yes":
        break
```
